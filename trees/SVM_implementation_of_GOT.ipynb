```python
x = df.drop(columns = "isAlive")
y = df.isAlive
```

```python
#train test and split
x_train , x_test , y_train ,y_test = train_test_split(x,y,test_size = 0.2 ,stratify = y)
```

```python
c3 =[x_train , x_test , y_train ,y_test]
for i in range (0,4):
  print(c3[i].shape)
```

```python
# importing svm 
from sklearn.svm import SVC
```

```python
#training the data using SVM
classifier = SVC()
classifier
classified = classifier.fit(x_train,y_train)
```

```python
#predicting the test spilt
pred = classifier.predict(x_test)
```

```python
#checking the accuracy
acc = accuracy_score(y_test,pred)
print(acc)
```

```python
# creating a function to build svm model predict the data and give accuracy also
def svm_class(x_train,y_train,x_test,y_test,kernel,evalue,gammavalue):
  mod = SVC(kernel = kernel,C = evalue,gamma = gammavalue)
  model = mod.fit(x_train,y_train)
  pred = mod.predict(x_test)
  acc = accuracy_score(y_test,pred)
  print("Accuracy is ",acc)
```

```python
c = [0.1 ,1,10,100]
c4 = [1,10,100]
for i in range (0,len(c4)):
  for j in range (0,len(c)):
    svm_class(x_train,y_train,x_test,y_test,'rbf',c4[i],c[j])
```
