DETERMING THE WINE QUALITY USING DECSION TREE 

```python
#importing the necessary lilbraries for handeling datasets and visualising them
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# assigning a object to contain the data set WineQT
df = pd.read_csv("WineQT.csv")
df.head()
```

```python
# getting the basic information on the dataset
print(df.info())
```

```python
# Our dependent variables is the column quality and the inpdepemdent ones are the features
# check for null values
df.isnull().sum()
# there are no null values
```

```python
# the id column is unnecessary for us and that is being removed
table = df.drop(columns = "Id")
```

Visualising the Data set


```python
sns.countplot(df.quality)
```

```python
#creating a column vector containing the index of the column name of the dataset
col = table.columns
col
```

```python
def plot(x):
  sns.scatterplot(df[x],df.quality)
  plt.title(x + ' vs quality of wine')
  plt.show()
  sns.distplot(df[x])
  plt.title("Distrinution of " + x)
  plt.show()
# defining a function to get the scatterplots of the features with respect to quality
```

```python
for i in range(0,len(col)-1):
  plot(col[i])
```

```python

table.quality.value_counts()
table.quality = table.quality.replace({3:5,4:5,7:8})

table.quality.value_counts()
```

```python
def clean_out(i):
  q1 = df[i].quantile(0.25)
  q2 = df[i].quantile(0.50)
  q3 = df[i].quantile(0.75)
  inter = q3 -q1
  u = q3 + 1.5*inter
  l = q1 - 1.5*inter
  for j in df[i]:
    if j>u or j<l:
      df[i] = df[i].replace(j,q2)
```

```python
for j in range(0,len(col)-1):
  clean_out(col[j])
```

Implementing Machine Learning Models

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# hyper parameter tuning
from sklearn.model_selection import GridSearchCV

params = {
    "criterion":("gini", "entropy"),  
    "max_depth":[5,10,15,20], 
    "min_samples_split":[2, 3, 4],"min_samples_split" :[2, 5, 10]
}
```

```python
# splitting the data set
x = table.drop(columns = "quality")
y = table.quality
```

```python
d_tree = GridSearchCV(DecisionTreeClassifier(),params)
```

Decision Tree

```python
#fitting the model
x1,x2,y1,y2 = train_test_split(x,y,test_size = 0.2, stratify = y,random_state=3)
model = d_tree.fit(x1,y1)
d_pred = d_tree.predict(x2)#predicting for test dataset
ac1 = accuracy_score(y2,d_pred)#checking the accuracy

print(ac1)
```

```python
print("The score of the tree is:",d_tree.score(x2,y2)*100)
```

```python
#checking best param
print(d_tree.best_params_)
```

```python
#importing the tree using tree_plot
from sklearn import tree
plt.figure(figsize=(60,40))
p = tree.plot_tree(DecisionTreeClassifier().fit(x1,y1),rounded= True)
```

```python
#creating classification table
d = classification_report(y2,d_pred)
print("Classification table of testing data\n",d)
d1 = classification_report(y1,model.predict(x1))
print("Classification table of training data\n",d1)
```
