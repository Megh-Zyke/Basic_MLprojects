Manipulating the data

```python
print(df.describe())
```

```python
df.info()
```

```python
print(df["fueltype"].value_counts())

df["fueltype"] = df["fueltype"].replace({"gas":0,"diesel":1})
```

```python
print(df["aspiration"].value_counts())

df["aspiration"] = df["aspiration"].replace({"std":0,"turbo":1})
```

```python
print(df["doornumber"].value_counts())

df["doornumber"] = df["doornumber"].replace({"two":0,"four":1})
```

```python
print(df["carbody"].value_counts())

df["carbody"] = df["carbody"].replace({"sedan":0,"hatchback":1,"wagon":2,"hardtop":3,"convertible":4})
```

```python
print(df["drivewheel"].value_counts())
df["drivewheel"] = df["drivewheel"].replace({"fwd":0,"rwd":1,"4wd":2})
```

```python
print(df["enginelocation"].value_counts())
df["enginelocation"] = df["enginelocation"].replace({"front":0,"rear":3})
```

```python
print(df["enginetype"].value_counts())
df["enginetype"] = df["enginetype"].replace({"ohc":0,"ohcf":1,"ohcv":2,"dohc":3,"l":4,"rotor":5,"dohcv":6})
```

```python
print(df["cylindernumber"].value_counts())
df["cylindernumber"] = df["cylindernumber"].replace({"four":4,"six":6,"five":5,"eight":8,"two":2,"three":3,"twelve":12})
```

```python
print(df["fuelsystem"].value_counts())
df["fuelsystem"]  = df["fuelsystem"].replace({"mpfi":0,"2bbl":1,"idi":2,"1bbl":3,"spdi":4,"4bbl":5,"mfi":6,"spfi":7})
```

```python
df.info()
```

Getting the data ready for ml

```python
df.head()
```

```python
x = df.drop(columns=["car_ID","CarName","price"])
y = df.price
```

```python
print(x.shape , y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state = 3)
```

```python
from sklearn.linear_model import LinearRegression as l_reg
```

```python
model = l_reg()
```

```python
model.fit(x_train,y_train)
```

```python
X = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
acc = r2_score(y_test,X)
print(acc)
print("The accuracy of the testing data is ",acc)


x2 = model.predict(x_train)
acc2 = r2_score(y_train,x2)
print("The accuracy of the training data is ",acc2)
```

```python
fig, (ax1,ax2)= plt.subplots(1,2,figsize = (15,10))

ax1.scatter(y_train,x2,c = "g")
ax1.plot(y_train,y_train, c = "b")
ax1.set_title('Linear Regression for training data')
ax1.set(xlabel='Orginal Data', ylabel='Predicted Data')

ax2.scatter(y_test,X,c = "r")
ax2.plot(y_test,y_test, c = "y")
ax2.set_title('Linear Regression for training data')
ax2.set(xlabel='Orginal Data', ylabel='Predicted Data')
```
