Importing Machine Learning Model

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:,0:4]
y = df.Target
```

```python
x.head()
```

```python
y.head()
```

```python
x1 , x2 ,y1 ,y2 = train_test_split(x,y,test_size=0.3, stratify = y , random_state = 1)
```

```python
model.fit(x1,y1)
```

```python
model.score(x2,y2)
```

```python
from sklearn.metrics import roc_auc_score , roc_curve
```

```python
fpr , tpr , threshold = roc_curve(model.predict(x2),y2)
```

```python
roc_auc_score(model.predict(x2),y2)
```

```python
plt.plot(fpr,tpr , linestyle = "dotted")
plt.plot([0,1],[0,1])
plt.title("ROC Curve of Logistic Regression model of dataset")
plt.xlabel("FPR    AUC SCORE " + str(roc_auc_score(model.predict(x2),y2)*100))
plt.ylabel("TPR")
```

```python
from sklearn.metrics import confusion_matrix
```

```python
matrix_ = confusion_matrix(y2,model.predict(x2))
sns.heatmap(matrix_,annot=True , cmap = "YlGnBu")
```
