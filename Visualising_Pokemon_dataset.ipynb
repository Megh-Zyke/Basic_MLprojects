```python
# declaring object df to contain the dataset pokemon.csv
df = pd.read_csv("pokemon.csv")
```

```python
df.head()
```

```python
table = df.drop(columns = "#")
```

```python
table.head()
```

```python
print("The revised information on pokemon.csv dataset is \n")
print(table.info())
print("The columns in this dataset are :",table.columns)
```

```python
table.describe()
```

```python
# renaming columns
table = table.rename(columns = {"Type 1":"Type1","Type 2":"Type2"})
```

```python
# getting some specific values
Grass_poison = table.query('Type1=="Grass" & Type2=="Poison"')
Grass_poison
# this table contains all the obejcts with type 1 as grass anf type 2 as poison
```

```python
fire_flying_legend = table[(table.Type1=="Fire")&(table.Type2=="Flying")&(table.Legendary== True)]
fire_flying_legend
# this table contains the pokemon with type 1 as fire type 2 as flying and are legendary
```

```python
plt.figure(figsize=(10,9))
p = sns.swarmplot(x = table.Legendary,y =table.Type1)
plt.show()
```

```python
# some countplots and comparison
fig , axs = plt.subplots(2,3,figsize =(25,8))
sns.countplot(gen1.Legendary ,ax=axs[0,0])
sns.countplot(gen1.Type1,ax=axs[0,1])
sns.countplot(gen1.Type2,ax=axs[0,2])
sns.countplot(table.Generation ,  hue=table.Legendary,ax=axs[1,0])
sns.countplot(table.Type1 , hue=table.Legendary ,ax=axs[1,1])
sns.countplot(table.Type2 , hue=table.Legendary  ,ax=axs[1,2])
```

```python
# plotting the type 1 pokemon with respect to Type 2
plt.figure(figsize=(22,12))
sns.countplot(table.Type1,hue = table.Type2 )
plt.show()
```
