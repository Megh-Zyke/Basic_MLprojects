Training and Testing Data

```python
# importing libraries
from sklearn.model_selection import train_test_split
```

```python
x = final.drop(columns = ["Type"])
y = final.Type
```

```python
x1,x2,y1,y2 = train_test_split(x,y,test_size = 0.3 , stratify = y ,random_state =20)
```

```python
print(x1.shape)
print(y1.shape)
print(x2.shape)
print(y2.shape)
```

IMPORTING MODELS

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression # importing logistic regression
model1 = LogisticRegression() # setting the model
model1.fit(x1,y1) #fitting the model
```

```python
y_pred = model1.predict(x2)
y_pred
```

```python
#checking accuracy
from sklearn.metrics import accuracy_score , confusion_matrix
acc = accuracy_score(y2,y_pred)
print("The accuracy is ",acc*100)
```

```python
matrix = confusion_matrix(y2,y_pred)
sns.heatmap(matrix , cmap= 'Blues')
plt.axes().set_title('Confusion Matrix')
```

```python
#plotting roc curve
from sklearn.metrics import roc_auc_score,roc_curve
```

```python
f1= pd.get_dummies(y_pred)
f2 = pd.get_dummies(y2)
```

```python
a,b,c,d,e = f1[0],f1[1],f1[2],f1[3],f1[4]
a1,b1,c1,d1,e1 = f2[0],f2[2],f2[2],f2[3],f2[4]
```

```python
auc1 = roc_auc_score(a,a1)
auc2 = roc_auc_score(c,c1)
auc3 = roc_auc_score(d,d1)
auc4 = roc_auc_score(e,e1)
auc_av = (auc1+auc3+auc4+auc5)/4
auc_av
```

```python
f1 , p1, t1 =roc_curve(a1,a)
f2 , p2, t2 =roc_curve(b1,b)
f3 , p3, t3 =roc_curve(c1,c)
f4 , p4, t4 =roc_curve(d1,d)
f5 , p5, t5 =roc_curve(e1,e)
```

```python
plt.figure(figsize=(10,10))
plt.plot(f1,p1,f3,p3,f4,p4,f5,p5 , linestyle = "dotted")
plt.plot([0,1],[0,1])
plt.title("ROC curve")
plt.xlabel("FPR   AUC SCORE = "+str((auc_av)*100))
plt.ylabel("TPR")
plt.legend("12345")
```
